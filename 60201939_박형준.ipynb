```python
def add_data(friend):
    
    katok.append(None)
    kLen=len(katok)
    katok[kLen-1]=(friend)
    for i in range (0,kLen):
        for j in range (0,kLen):
            if katok[i][1]>katok[j][1]:
                a= katok[i]
                katok[i]=katok[j]
                katok[j]=a
            j+=1
        i+=1
            

                
                
def delete_data(position):
    if position<0 or position>len(katok):
        print("   .")
        return
    kLen=len(katok)
    katok[position]=None
    
    for i in range(position+1,kLen):
        katok[i-1]=katok[i]
        
    del(katok[kLen-1])
    
katok=[]
select=-1


if __name__=="__main__":
    
    while (select !=3):
        select= int(input(" (1:,2:,3:)-->"))
        
        if (select==1):
            data= input(" --> ")
            data2=int(input(" -->"))
            sum =(data,data2)
            
            add_data(sum)
            print(katok)
                    
               
        elif(select==2):
            pos=int(input(" -->"))
            delete_data(pos)
            print(katok)
        elif(select==3):
            print(katok)
            exit
        else:
            print("1~3  ")
            continue
```
